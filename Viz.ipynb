<h3> Plot of the number of articles for each month during the year. Each line represents a year. 

```python
year_month_grouped = df.groupby(['year', 'month']).count()

for year in range(1996, 2019):
    year_month_grouped.loc[year]['day'].plot.line()

plt.title('Number of articles per month (Each line is a year)')
```

<h5> This shows the total articles by month allowing us to find which times had significant dips regarding the number of articles

```python
year_month_grouped['day'].plot.line()
plt.title('Number of articles per month across all years')
```
